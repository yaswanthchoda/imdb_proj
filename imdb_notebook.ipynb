# IMDB Notebook

This is a simple notebook which is used to get movies information from IMDB site.Here I am using imdb library provided by python to get top 250 movies data from IMDB and displaying scatter plots using this data.

### Steps to execute this notebook

- Goto kernal option and select 'imdb_env'
- press Ctrl+enter to execute cell

The following cell Imports imdb nd get data for top 250 movies

```python
# importing the module
import imdb
import numpy as np
import matplotlib.pyplot as plt

# creating instance of IMDb
im_inst = imdb.IMDb()

# getting top 250 movies
top_movies_list = im_inst.get_top250_movies()

# print(type(search))
title = []
rating = []
votes = []
for movie in top_movies_list[:]:
    title.append(movie.data['title'])
    rating.append(movie.data['rating'])
    votes.append(movie.data['votes'])
```

### Create Scatter plots with rating and votes

```python
plt.scatter(rating, votes)
plt.xlabel('Rating')
plt.ylabel('Votes')
plt.savefig('scatter_plot1.png', dpi=100)
plt.show()
```

### Create Scatter plots with names using ratings and votes

please update the cnt value with intger less than 250 to display the number of movies want to create the scatter plots

```python
cnt = 10
fig, ax = plt.subplots()
ax.scatter(rating[:cnt], votes[:cnt])
for i, txt in enumerate(title[:cnt]):
    ax.annotate(txt, (rating[i], votes[i]))
    
fig.savefig('scatter_plot2.png', dpi=100)
```
